# Data Cleaning
## Filling the columns whose missing values are less than 10%
### We use the central tendency

```python
#Filling of missing values in MinTemp column(< than 10% so fill with central tendency(mean))
MinTemp_mean = Data["MinTemp"].mean()
Data["MinTemp"] = Data["MinTemp"].fillna(MinTemp_mean)
Data["MinTemp"]
```

```python
# Checking is it any null value in MinTemp column 
Data["MinTemp"].isna().sum()
```

```python
#Filling of missing values in MaxTemp column(< than 10% so fill with central tendency(mean))
MaxTemp_mean = Data["MaxTemp"].mean()
Data["MaxTemp"] = Data["MaxTemp"].fillna(MaxTemp_mean)
Data["MaxTemp"]
```

```python
Data["MaxTemp"].isna().sum()

```

```python
#Filling of missing values in WindGustSpeed column(< than 10% so fill with central tendency(mean))
WindGustSpeed_mean = Data["WindGustSpeed"].mean()
Data["WindGustSpeed"] = Data["WindGustSpeed"].fillna(WindGustSpeed_mean)
Data["WindGustSpeed"]
```

```python
Data["WindGustSpeed"].isna().sum()
```

```python
WindSpeed9am_mean = Data["WindSpeed9am"].mean()
Data["WindSpeed9am"] = Data["WindSpeed9am"].fillna(WindSpeed9am_mean)
Data["WindSpeed9am"]
```

```python
WindSpeed3pm_mean = Data["WindSpeed3pm"].mean()
Data["WindSpeed3pm"] = Data["WindSpeed3pm"].fillna(WindSpeed3pm_mean)
Data["WindSpeed3pm"]
```

```python
Humidity9am_mean = Data["Humidity9am"].mean()
Data["Humidity9am"] = Data["Humidity9am"].fillna(Humidity9am_mean)
Data["Humidity9am"]
```

```python
Humidity3pm_mean = Data["Humidity3pm"].mean()
Data["Humidity3pm"] = Data["Humidity3pm"].fillna(Humidity3pm_mean)
Data["Humidity3pm"]
```

```python
Temp9am_mean = Data["Temp9am"].mean()
Data["Temp9am"] = Data["Temp9am"].fillna(Temp9am_mean)
Data["Temp9am"]
```

```python
Temp3pm_mean = Data["Temp3pm"].mean()
Data["Temp3pm"] = Data["Temp3pm"].fillna(Temp3pm_mean)
Data["Temp3pm"]
```
